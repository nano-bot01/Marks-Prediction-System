## Importing Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## Load Dataset

```python
dataset = pd.read_csv("student_info.csv", header=0)
```

```python
df = dataset.fillna(dataset.mean())
```

```python
df.isnull().sum()
```

## Visualization of data

```python
dataset = df
```

```python
plt.figure(figsize = (16,9))
plt.scatter(dataset.study_hours, dataset.student_marks, color = 'g' )
plt.xlabel("Hours of Study")
plt.ylabel("Student marks")
plt.title("Scatter Plot between Hours of Study and Marks of Student", fontsize = 18)
plt.grid(True)  # Add grid to the plot
plt.show()
```

```python
plt.figure(figsize = (50,15))
sns.countplot(x=dataset["study_hours"])

# distribution of hour of study
```

```python
plt.figure(figsize = (60,20))
sns.countplot(x=dataset["student_marks"])

# distribution of marks
```

```python
sns.pairplot(dataset, hue= 'student_marks',vars = ['study_hours', 'student_marks'])

# pair plot in dataset of outcome with all columns
```

```python
X = dataset.drop('student_marks', axis = 'columns')
Y = dataset.drop('study_hours', axis = 'columns')

print("Shape of X = : ", X.shape)
print("Shape of Y = : ", Y.shape)
```

## Train-Test Split

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 51)
```

```python
print("Shape of X_train : ", X_train.shape)
print("Shape of X_test : ", X_test.shape)
print("Shape of Y_train : ", Y_train.shape)
print("Shape of Y_test : ", X_test.shape)
```

```python
lr = LinearRegression()

lr.fit(X_train,Y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

### Prediction

```python
lr.predict([[8]])[0][0].round()
```

```python
Y_pred = lr.predict(X_test)
Y_pred
```

```python
pd.DataFrame(np.c_[X_test,Y_test, Y_pred], columns = ['study_hours', 'student_marks_orginal','student_marks_predicted'])
```

## Fine Tue your model

```python
lr.score(X_test, Y_test)
```

```python
plt.figure(figsize = (16,9))
plt.scatter(X_train,Y_train)
plt.plot(X_train, lr.predict(X_train), color = 'r')
plt.show()
```

## Saving the model 

```python
import pickle 
# importing the library

filename = "Marks Prediction trained_model.pkl"
pickle.dump(lr, open(filename, 'wb'))
# saving file 
```
